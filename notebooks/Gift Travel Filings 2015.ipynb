```python
import agate
```

```python
gift_travel = agate.Table.from_csv('../data/2015Travel.txt', delimiter='\t')
```

```python
gift_travel.print_table()
```

```python
print(gift_travel)
```

```python
#Which state's reps account for most gift travel?

by_state = gift_travel.group_by('State')
```

```python
by_state_count = by_state.aggregate([
        ('count', agate.Count())
    ])

by_state_order = by_state_count.order_by('count', reverse=True)
```

```python
by_state_order.print_bars('State', 'count', width=50)
```

```python
#Who are the reps who filed the most claims?

by_member = gift_travel.group_by('MemberName')

by_member_count = by_member.aggregate([
        ('count', agate.Count())
                                      ])
by_member_count.print_table()
```

```python
member_order = by_member_count.order_by('count', reverse=True)

member_order.print_table()

member_order.print_bars('MemberName', 'count', width=50)
```

```python
top_10_filers = member_order.limit(10)

top_10_filers.print_table()
```

```python
#What's the most popular destination?

destinations = gift_travel.group_by('Destination')

destinations_count = destinations.aggregate([
        ('count', agate.Count())
    ])

top_destinations = destinations_count.order_by('count', reverse=True)

top_destinations.print_table()

top_destinations.print_bars('Destination', 'count', width=50)
```

```python
#Who is paying for these trips?

sponsors = gift_travel.group_by('TravelSponsor')

sponsors_count = sponsors.aggregate([
        ('count', agate.Count())
    ])

top_sponsors = sponsors_count.order_by('count', reverse=True)

top_sponsors.print_table()
top_sponsors.print_bars('TravelSponsor', 'count', width=50)
```
